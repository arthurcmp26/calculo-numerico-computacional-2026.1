```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**3 - x - 2

x = np.linspace(-3, 3, 400)
y = f(x)

plt.figure()
plt.axhline(0)  # eixo x
plt.plot(x, y)
plt.title("Gráfico de f(x) = x³ - x - 2")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid()
plt.show()
```

### Quantas Raízes aparentes existem ?

Existe uma única raíz real aparente no intervalo [-3, 3].

### Em que região aproximada ela está ?

Observando o gráfico, a raíz parece estar entre 1 e 2, mais próximo de 1,5.

```python
for i in range(-3, 4):
    print(f"x = {i}, f(x) = {f(i)}")
```

### Em qual intervalo ocorre mudança de sinal ?

A mudança de sinal ocorre no intervalo de "f(1) = -2" e "f(2) = 4".

### Existe mais de um intervalo válido ?

Não, apenas entre "f(1)" e "f(2)" ocorre a mudança de sinal

### Qual intervalo você escolheria para estudar primeiro ?

O intervalo (1,2) pois é o único que garante a existência da raíz, passando pelo zero.

```python
a = 1
b = 2

for i in range(5):  # 5 iterações
    m = (a + b) / 2
    print(f"Iteração {i+1}:")
    print(f"a = {a}, b = {b}, m = {m}, f(m) = {f(m)}\n")

    if f(a) * f(m) < 0:
        b = m
    else:
        a = m
```

### O intervalo pode ser reduzido ?

Sim. Sempre escolhemos a metade onde ocorre mudança de sinal, reduzindo o intervalo progressivamente.

### A raiz parece estar mais próxima de qual extremo ?

Após algumas iterações, observa-se que a raiz está aproximadamente em:

x ≈ 1,52

Ela fica mais próxima de 1,5 do que de 2, portanto mais próxima do extremo inferior inicialmente.

### Que tipo de método estamos construindo intuitivamente ?

Estamos construindo intuitivamente o Método da Bisseção, que é um método numérico de localização de raízes baseado em mudança de sinal.

### **Análise Conceitual**

### 1) Por que mudança de sinal indica raiz ?

Porque pelo Teorema do Valor Intermediário, se uma função contínua
muda de sinal em um intervalo, então existe pelo menos uma raiz nesse intervalo.


### 2) Ausência de mudança de sinal garante ausência de raiz ?

Não. Pode existir raiz de multiplicidade par (ex: (x-1)^2),
onde a função toca o eixo x mas não muda de sinal.


### 3) Por que localizar antes de aplicar método iterativo ?

Porque métodos como bisseção precisam de um intervalo
válido com garantia de raiz para convergir corretamente.

### **Desafio** **Extra**

```python
def g(x):
    return (x - 1)**2

for i in range(-1, 4):
    print(f"x = {i}, g(x) = {g(i)}")
```

### Existe raíz ?

Sim. Existe raíz em "x = 1"

### Há mudança de sinal ?

Não. A função apenas toca o eixo x e volta a ser positiva.
Ela nunca assume valores negativos.

### Um método baseado apenas em sinal funcionária ?

Não, pois como não há mudança de sinal, o método da bisseção não detectaria essa raiz.

Isso ocorre porque é uma raiz de multiplicidade par.
