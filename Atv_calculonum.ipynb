## Questão 1
O resultado de "math.sqrt(2)**2" não é exatamente 2 devido às limitações da representação de números reais em ponto flutuante.
O valor de √2 é irracional e precisa ser aproximado pelo computador, o que gera pequenos erros numéricos.

## Questão 2
