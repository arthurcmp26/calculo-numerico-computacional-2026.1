# **Implementação em Python (Gauss)**

```python
import numpy as np

def eliminacao_gauss(A, b):
    A = A.astype(float)
    b = b.astype(float)
    n = len(b)

    # Etapa 1: Eliminação (forma triangular)
    for k in range(n - 1):
        if A[k][k] == 0:
            raise ValueError("Pivô zero encontrado! Pode precisar de pivoteamento.")

        for i in range(k + 1, n):
            fator = A[i][k] / A[k][k]
            A[i, k:] = A[i, k:] - fator * A[k, k:]
            b[i] = b[i] - fator * b[k]

    # Etapa 2: Substituição regressiva
    x = np.zeros(n)

    for i in range(n - 1, -1, -1):
        if A[i][i] == 0:
            raise ValueError("Sistema pode não ter solução única.")

        soma = np.dot(A[i, i+1:], x[i+1:])
        x[i] = (b[i] - soma) / A[i][i]

    return x
```

# **Teste**

```python
A = np.array([
    [2, 1, -1],
    [-3, -1, 2],
    [-2, 1, 2]
])

b = np.array([8, -11, -3])

# Seu método
x_gauss = eliminacao_gauss(A.copy(), b.copy())

# Comparação com numpy
x_numpy = np.linalg.solve(A, b)

print("Solução (Gauss):", x_gauss)
print("Solução (NumPy):", x_numpy)
```

# **Análise**

## O resultado coincide com o numpy?

Sim, Os valores obtidos pelo método de Gauss coincidem com os do numpy.linalg.solve, podendo ter apenas pequenas diferenças devido a arredondamento.

## Qual etapa foi mais difícil?
A etapa de eliminação, pois exige cuidado com os índices, cálculos do fator e atualização correta da matriz e do vetor.

## O método funciona sempre?
Não, Ele pode falhar quando há pivô zero ou quando o sistema não possui solução única.

# **Desafio**

```python
A2 = np.array([
    [1, 1, 1],
    [2, 3, 1],
    [1, 2, 3]
])

b2 = np.array([6, 10, 13])

x2_gauss = eliminacao_gauss(A2.copy(), b2.copy())
x2_numpy = np.linalg.solve(A2, b2)

print("Gauss:", x2_gauss)
print("NumPy:", x2_numpy)
```

# **Investigação (Sistema Problemático)**

```python
A3 = np.array([
    [1, 1],
    [2, 2]
])

b3 = np.array([2, 4])

try:
    x3 = eliminacao_gauss(A3.copy(), b3.copy())
    print("Solução:", x3)
except Exception as e:
    print("Erro:", e)
```

## O que acontece com o algoritmo?

Ele encontra uma linha com zeros e não consegue continuar corretamente

## O sistema possui solução única?
Não, Possui infinitas soluções.

## Como identificar esse caso?
Quando aparece uma linha do tipo:
0x + 0y = 0, ou então quando o determinante da matriz é zero.

# **Reflexão Final**

## Por que o método de Gauss funciona?
Porque transforma o sistema original em um sistema equivalente mais simples, preservando as soluções.

## Onde sistemas lineares aparecem na engenharia?
Estruturas, Circuitos elétricos, Mecânica dos fluidos ou Análise de dados

## Qual a importância de resolver sistemas grandes?
Permite modelar problemas reais complexos, como simulações, otimizações e projetos de engenharia.
