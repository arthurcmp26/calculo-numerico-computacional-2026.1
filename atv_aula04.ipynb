```python
import numpy as np
import matplotlib.pyplot as plt

# Função
def f(x):
    return x**3 - 9*x + 3

# Intervalo maior (menos zoom)
x = np.linspace(-5, 5, 200)
y = f(x)

# Plot
plt.axhline(0)
plt.plot(x, y)
plt.title("f(x) = x³ - 9x + 3")
plt.grid()
plt.show()
```

###Em que região parece estar localizada a raiz?

A raiz está entre 0 e 1.

### Existe mais de uma raiz no intervalo analisado?

Não, especificamente no intervalo analisado não.

# Método da Bissecção

```python
def bisseccao(a, b, iteracoes=10):
    print("Iter |    a    |    b    |    c    |   f(c)")

    for i in range(iteracoes):
        c = (a + b) / 2

        print(f"{i+1:>4} | {a:.6f} | {b:.6f} | {c:.6f} | {f(c):.6f}")

        if f(a) * f(c) < 0:
            b = c
        else:
            a = c

    return c


raiz_biss = bisseccao(0, 1)
print("\nRaiz aproximada (Bissecção):", raiz_biss)
```

# Método de Newton

```python
def df(x):
    return 3*x**2 - 9


def newton(x0, iteracoes=10):
    print("Iter |    x_n    |   f(x_n)")

    x = x0

    for i in range(iteracoes):
        fx = f(x)
        dfx = df(x)

        print(f"{i+1:>4} | {x:.6f} | {fx:.6f}")

        if dfx == 0:
            print("Derivada zero! Parando.")
            break

        x = x - fx/dfx

    return x


raiz_newton = newton(raiz_biss)

print("\nRaiz aproximada (Newton):", raiz_newton)
```

# Comparação entre os métodos

## Número de iterações:
O método da bissecção precisa de um maior número de iterações para alcançar uma boa aproximação de raiz, enquanto em Newton converge em poucas iterações.

## Velocidade de Convergência:

A bissecção converge de forma mais lenta, enquanto o método de Newton converge de forma mais rápida.

## Facilidade de Implementação:

A bissecção é mais simples de implementar, pois não requer derivadas. Já o método de Newton é mais complexo, pois exige o cálculo da derivada da função.

# Experimento Extra

```python
newton(0)
newton(1)
newton(10)
```

## Newton sempre converge ?

Não, pois pode falhar se o chute inicial for ruim ou se a derivada for zero.

## O resultado depende do chute inicial ?

Sim, pois o valor inicial influencia a convergência e a raiz encontrada.

# Reflexão Final

### Por que o método de Newton costuma convergir mais rápido que a bissecção?

Porque utiliza a derivada para “ir direto” à raiz, tendo convergência mais rápida que a bissecção.

### Em quais situações o método da bissecção pode ser preferível?

Quando se deseja garantia de convergência ou quando não se conhece a derivada da função.

### Qual método você utilizaria em um problema real de engenharia?

Depende do caso. Newton para maior rapidez quando há boa estimativa inicial, bissecção quando se busca segurança.
