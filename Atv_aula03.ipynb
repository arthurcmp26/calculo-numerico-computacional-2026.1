# **Visualização do Gráfico**

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**3 - 9*x + 3

x = np.linspace(-3, 3, 1000)
y = f(x)

plt.figure()

plt.plot(x, y, label="f(x) = x³ - 9x + 3")
plt.axhline(0)
plt.axvline(0)

plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Gráfico da função f(x) = x³ - 9x + 3")
plt.legend()

plt.show()
```

### Onde parece estar localizada a raiz ?

A raiz está aparentemente entre 0 e 1.

### Existe mais de uma raiz ?

Apesar da função ser cúbica e poder possuir até três raízes reais, o método da bissecção foi aplicado apenas no intervalo [0, 1]. No entanto, ele encontrou somente a raiz que está localizada nesse intervalo.

# **Verificação** **de** **Mudança** **de** **Sinal**

```python
def f(x):
    return x**3 - 9*x + 3

a = 0
b = 1

print("f(a) =", f(a))
print("f(b) =", f(b))

if f(a)*f(b) < 0:
    print("Existe mudança de sinal. O intervalo é válido.")
else:
    print("Não existe mudança de sinal.")
```

### Existe mudança de sinal ?

Sim.

### O intervalo escolhido é adequado ?

Sim, devido a mudança de sinal

# **Método** **da** **Bissecção**

```python
def f(x):
    return x**3 - 9*x + 3

def bisseccao(a, b, iteracoes):

    if f(a)*f(b) >= 0:
        print("Intervalo inválido")
        return

    print("Iteração | a | b | c | f(c)")

    for i in range(iteracoes):

        c = (a + b) / 2

        print(i+1, a, b, c, f(c))

        if f(a)*f(c) < 0:
            b = c
        else:
            a = c

bisseccao(0, 1, 10)
```

# **Tabela de Iterações**

```python
import pandas as pd

def f(x):
    return x**3 - 9*x + 3

def bisseccao_tabela(a, b, iteracoes):

    dados = []

    for i in range(1, iteracoes + 1):

        c = (a + b) / 2

        dados.append([i, a, b, c, f(c)])

        if f(a)*f(c) < 0:
            b = c
        else:
            a = c

    tabela = pd.DataFrame(dados, columns=["Iteração","a","b","c","f(c)"])

    return tabela


tabela = bisseccao_tabela(0, 1, 10)

print(tabela)
```

# **Análise** **de** **Convergência**

### 1. Quantas iterações foram necessárias para obter uma boa aproximação da raiz?

Foram realizadas 10 iterações, o que forneceu uma boa aproximação de raiz (0.336914).

### 2. O intervalo diminuiu pela metade a cada iteração?

Sim, pois no método da bissecção, a cada iteração o intervalo de onde a raiz está localizada é dividido pela metade.

### 3. O resultado obtido parece coerente com o gráfico da função?

Sim, pois o gráfico indicava que a raiz estava localizada entre 0 e 1, e o método encontrou uma aproximação de 0.336914, que está dentro deste intervalo.

# **Reflexão** **Final**

### 1. Por que o método da bissecção sempre converge quando há mudança de sinal ?

Porque quando a função muda de sinal dentro de um intervalo, isso significa que ela cruza o eixo x em algum ponto ali. O método da bissecção aproveita isso dividindo o intervalo ao meio várias vezes e sempre escolhendo a parte onde ainda existe mudança de sinal. Assim, a cada passo a gente vai se aproximando cada vez mais da raiz.

### 2. Qual é a principal limitação desse método ?

A principal limitação é que ele pode ser um pouco lento, já que a cada iteração o intervalo só diminui pela metade. Ele também só funciona quando conseguimos encontrar um intervalo onde a função muda de sinal.

### 3. Em quais situações ele pode ser mais adequado que outros métodos ?

Ele é mais adequado quando queremos ter certeza de que o método vai funcionar, mesmo que demore um pouco mais. Também é útil quando não sabemos exatamente onde está a raiz, mas conseguimos identificar um intervalo onde a função muda de sinal.
