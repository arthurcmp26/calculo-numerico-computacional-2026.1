```python
import math

def f(x):
    return math.cos(x) - x

def bisseccao(f, a, b, iteracoes):

    for i in range(iteracoes):

        c = (a + b) / 2

        print("Iteração:", i+1,
              "a =", a,
              "b =", b,
              "c =", c,
              "f(c) =", f(c))

        if f(a) * f(c) < 0:
            b = c
        else:
            a = c

    return c


raiz = bisseccao(f, 0, 1, 10)

print("\nRaiz aproximada:", raiz)
```

# O que aconteceu ?

A cada iteração o intervalo foi dividido ao meio e o método escolheu a metade onde a mudança de sinal continuava ocorrendo. Assim, o valor de c foi se aproximando cada vez mais da raiz da função.

```python
import pandas as pd
import math

def f(x):
    return math.cos(x) - x

def bisseccao_tabela(a, b, iteracoes):

    dados = []

    for i in range(1, iteracoes + 1):

        c = (a + b) / 2

        dados.append([i, a, b, c, f(c)])

        if f(a) * f(c) < 0:
            b = c
        else:
            a = c

    tabela = pd.DataFrame(dados, columns=["Iteração","a","b","c","f(c)"])

    return tabela


tabela = bisseccao_tabela(0, 1, 10)

print(tabela)
```

```python
print("5 iterações:")
print("Raiz aproximada:", bisseccao(f, 0, 1, 5))

print("\n10 iterações:")
print("Raiz aproximada:", bisseccao(f, 0, 1, 10))

print("\n20 iterações:")
print("Raiz aproximada:", bisseccao(f, 0, 1, 20))
```

# O que acontece com o resultado ?

À medida que o número de iterações aumenta, o valor encontrado para a raiz fica cada vez mais preciso. O intervalo onde a raiz está localizada vai diminuindo progressivamente, fazendo com que o valor calculado se aproxime cada vez mais da raiz real da função.
