```python
def f(x):
    return x**3 - x - 2

def secante(x0, x1, tol=1e-6, max_iter=100):
    print(f"{'Iter':<5} {'x0':<10} {'x1':<10} {'x2':<12} {'Erro':<12}")

    for i in range(max_iter):
        if f(x1) - f(x0) == 0:
            print("Divisão por zero!")
            return None

        x2 = x1 - f(x1)*(x1 - x0)/(f(x1) - f(x0))
        erro = abs(x2 - x1)

        print(f"{i:<5} {x0:<10.6f} {x1:<10.6f} {x2:<12.6f} {erro:<12.6f}")

        if erro < tol:
            print(f"\nRaiz aproximada: {x2}")
            return x2

        x0, x1 = x1, x2

    print("Não convergiu")
    return None


secante(1, 2)
```

## Qual método convergiu mais rápido ?

Newton converge mais ráoido, mas também vai depender do chute inicial.

## A secante sempre converge ?

Não, pois ela pode falhar dependendo dos valores iniciais ou da função. Não é garanttida como da bissecção.

## Newton depende do chute inicial ?

Sim, pois se o chute for ruim pode não convergir, convergir para outra raiz, ou até divergir.
